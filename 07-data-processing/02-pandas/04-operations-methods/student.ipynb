```python
import pandas as pd

data = {
    'Product_ID': [101, 102, 103, 104, 105],
    'Product_Name': ['Laptop', 'Smartphone', 'Tablet', 'Headphones', 'Smartwatch'],
    'Category': ['Electronics', 'Electronics', 'Electronics', 'Accessories', 'Electronics'],
    'Price': [1200, 800, 400, 100, 300],
    'Quantity_Sold': [50, 100, 80, 200, 120]
}

products = pd.DataFrame(data)
products
```

```python
products['Revenue'] = products['Price'] * products['Quantity_Sold']

print(products[['Product_Name', 'Revenue']])
print("total revenue:")
print(products['Revenue'].sum())
```

```python
print(products['Price'].mean())
print(products['Price'].median())
print()
print(products['Quantity_Sold'].mean())
print(products['Quantity_Sold'].median())
```

```python
products_concat = pd.concat([products, products], axis=1)
products_concat
```

```python
products.loc[2, 'Price'] = pd.NA
products.loc[3, 'Quantity_Sold'] = pd.NA

numeric_columns = ['Price', 'Quantity_Sold']
products[numeric_columns] = products[numeric_columns].fillna(products[numeric_columns].mean())
products
```
